A Logistic regression classifier consists of 3 distinct parts:
1. A multivariate linear function :  $z = \mathbf{w}^\top\mathbf{x} = \sum_i w_i*x_i +w_0$ , where $\mathbf{w}$ are the model parameters (including the bias term $w_0$) and $\mathbf{x}$ the homogeneous coordinates of the input

2. The sigmoid activation function $\sigma(z)=\frac{1}{1+\text{e}^{(-z)}}$ 

3. A binary step function (thresholding) at $\sigma(z)>0.5$.

**Hint**
`z`
 must be zero in:

$$\sigma(z)=\frac{1}{1+\text{e}^{(-z)}}$$
for $\sigma(z)=0.5$, hence solve $w_0 + w x = 0$


---

```python
def linear_sigmoid(x,b,w) :  # sigmoid function
    """
    :param x: 1D array of the (single) input-feature values.
    :param b: The bias parameter of the model.
    :param W: The weight parameter of the model.
    
    :return (float): output values of the sigmoid function. 
    """
    # Linear combination: z = w0 + w1 * x
    z = b + w * x
    
    # Apply sigmoid function: sigma(z) = 1 / (1 + exp(-z))
    return 1 / (1 + np.exp(-z))

xs = np.linspace(-100,100,1000) ### linspace of "input features"

params = [6,0.80] # bias and weight

# Calculate the decision boundary where sigma(z) = 0.5
# We know that sigma(z) = 0.5 when z = 0, so solve for x where b + w * x = 0
b, w = params
Decision_boundary = -b / w  # Solves for x at which z = 0 (sigma(z) = 0.5)

### plotting ####
plt.figure(figsize=(8,6))
plt.plot(xs, linear_sigmoid(xs,params[0],params[1]),label = r"$\sigma(x)$ = 1/(1 + exp(-(%.2f + %.2fx))"%(params[0],params[1]))
plt.plot([Decision_boundary,Decision_boundary],[0,1],'--C3',label = 'decision boundary')
plt.title(r"Sigmoid $\sigma(x)$", fontsize=20)
plt.legend()
plt.show()
```

## Task 1 reflection

**What happens when you change bias?**

* Adding bias: The curve shifts to the left. This means that We accept lower x values, as being labeled "true"
* Reducing bias: The curve shifts to the right. This means that x values must be higher, to be labeled "true"

**What happens when you change w_1 "slope" ?**

* Adding to w_1: The curve becomes more narrow. When moving from the decision boundary, we are more certain they are either true/false
* Reducing w_1: The curve becomes more broad. When moving away from the decision boundary, the odds are not as close to 100 %

**Task 2 (medium): The prediction functionüë©‚Äçüíª**
1. In the cell below implement the function `predict`
  using the `linear_sigmoid`
 function and binary thresholding to predict the class given input-features and model parameters.


---

```python
def predict(x,w):
    """
    :param x: 1D array of the (single) input-feature values.
    :param w: The list of the model parameters, [bias, weight]. 
    
    :return: Boolean array same size as x, where a True values signifies class2, and False signifies class1
    """
    # Unpack model parameters
    bias, weight = w
    
    # Compute the sigmoid output for each input feature
    sigmoid_output = linear_sigmoid(x, bias, weight)
    
    #Threshold:
    # If sigmoid output >= 0.5, classify as class2 (True), otherwise as class1 (False)
    return sigmoid_output >= 0.5
```

```python
# ------For manually setting weights and bias-----
params = [0.303,0.432] # bias and weight
x_all = np.concatenate([x1,x2])
y_all = np.concatenate([y1+np.random.normal(0,0.01,200),y2+np.random.normal(-0,0.01,200)])
y_bool = predict(x_all,params)
plt.figure(figsize=(8,6))
plt.plot(x_all[~y_bool], y_all[~y_bool],'p',color='orange',label ='Predicted class1')
plt.plot(x_all[y_bool], y_all[y_bool],'go',label = 'Predicted class2')
plt.plot(xs, linear_sigmoid(xs,params[0],params[1]),'b',label = r"$\sigma(x)$ = 1/(1 + exp(-(%.2f + %.2fx))"%(params[0],params[1]))
plt.plot([-params[0]/params[1],-params[0]/params[1]],[-0.04,1.04],'--C3',label = 'decision boundary')
plt.title(r"Logistic regression prediction", fontsize=20)
plt.legend()
plt.show()

def accuracy(predictions, targets):
    """
    Calculate accuracy of predictions.
    
    :param predictions: Array of predicted classes for the data points.
    :param targets: Array of actual classes for the data points.
    
    :return: Fraction of correctly predicted points.
    """
    correct_predictions = np.sum(predictions == targets)
    return correct_predictions / len(targets)

# Since y_bool is boolean, convert it to 1 and 0 for comparison with y_all
predictions = y_bool.astype(int)

# Calculate the accuracy using the accuracy function
#Predictions vs targets.
acc = accuracy(predictions, y_all.round())  # Rounding y_all to 0 and 1 to match the predictions

# Output the accuracy
print(f'Accuracy of the classifier with parameters {params}: {acc:.3f}')


# ---- Using Grid search for params-----

# Define ranges for bias and weight parameters
bias_range = np.linspace(-10, 10, 100)     # Try values from 1 to 3 for bias in 5 steps
weight_range = np.linspace(0.01, 2, 100)  # Try values from 0.01 to 0.1 for weight in 5 steps

# Initialize variables to store the best parameters and highest accuracy
best_accuracy = 0
best_params = [0, 0]

# Perform grid search over all combinations of bias and weight
for bias in bias_range:
    for weight in weight_range:
        # Current set of parameters
        params = [bias, weight]
        
        # Predict classes using the current parameters
        y_bool = predict(x_all, params)
        predictions = y_bool.astype(int)  # Convert boolean predictions to integers
        
        # Calculate the accuracy with the current parameters
        acc = accuracy(predictions, y_all.round())
        
        # Update best parameters if current accuracy is higher
        if acc > best_accuracy:
            best_accuracy = acc
            best_params = params

# Output the best parameters and highest accuracy
print(f'Best parameters: {best_params}')
print(f'Highest accuracy: {best_accuracy:.3f}')

# Visualization using the best parameters
plt.figure(figsize=(8,6))
plt.plot(x_all[~y_bool], y_all[~y_bool], 'p', color='orange', label='Predicted class1')
plt.plot(x_all[y_bool], y_all[y_bool], 'go', label='Predicted class2')
plt.plot(xs, linear_sigmoid(xs, best_params[0], best_params[1]), 'b', label=r"$\sigma(x)$ = 1/(1 + exp(-(%.2f + %.2fx))" % (best_params[0], best_params[1]))
plt.plot([-best_params[0]/best_params[1], -best_params[0]/best_params[1]], [-0.04, 1.04], '--C3', label='Decision boundary')
plt.title(r"Logistic regression prediction", fontsize=20)
plt.legend()
plt.show()
```

**Task 4 (medium): Increasing input dimensionsüë©‚Äçüíªüí°**
In this task the logistic regression function is modified to take 2 features as input (x,y).
1. Complete the `sigmoid2D`
 function such that it can take a vector of two 1D-arrays as input.
2. Run the cell below to visualize a 3D-plot of the 2-variable sigmoid function and the decision boundary plane 
3. The decision boundary is defined by the plane where $z\in[0,1]$ and x,y such that $ b + w_1*x+w_2*y = 0 $ . Why is this?
4. How does changing the individual model parameters  $w_1$  and  $w_2$  affect the orientation and slope of the decision boundary plane in the 3D plot?

```python
def sigmoid2D(X,params) :  # sigmoid function
    """
    :param X: tuple of input features (x,y).
    :param parametes: List of the model parameters.
    :return: output values of the sigmoid function. 
    """
    #Unpacked input features
    x, y = X
    b, w_x, w_y = params  # Unpack model parameters
    
    # Linear combination: z = b + w_x * x + w_y * y
    z = b + w_x * x + w_y * y
    
    # Apply sigmoid function: sigma(z) = 1 / (1 + exp(-z))
    return 1 / (1 + np.exp(-z))

#### Plotting
x = np.linspace(-50, 50, 1000)
y = np.linspace(-50, 50, 1000)
X, Y = np.meshgrid(x, y)

params3 = [0.07,0.037,0.085] # bias, w_x, w_y

Z = sigmoid2D((X, Y),params3)

fig = plt.figure(figsize=(15,7))

# ----- First subplot: 3D sigmoid function -----
ax1 = fig.add_subplot(1, 2, 1, projection='3d')  # 1 row, 2 columns, first plot

# Decision boundary plane calculation
y_decision_boundary = -(params3[0]/params3[2]) - (params3[1]/params3[2]) * x

# Create mesh for decision boundary plane in 3D space
Y_decision, Z_decision = np.meshgrid(y_decision_boundary, np.linspace(0, 1, 1000))  # z from 0 to 1 for probabilities
X_decision = np.tile(x, (len(Y_decision), 1))  # x values remain constant for the plane

# Plot the decision boundary as a wireframe to make it clearly visible
boundary_surface = ax1.plot_wireframe(X_decision, Y_decision, Z_decision, color='black', alpha=0.6, label='Decision Boundary')
sigmoid_surface = ax1.plot_surface(X, Y, Z, color='green', alpha=0.3)  # making it more transparent

ax1.set_title(r"Sigmoid : $\sigma(x,y)$ = 1/(1 + exp(-(%.2f + %.2fx + %.2fy))"%(params3[0],params3[1],params3[2]), fontsize=20)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z (Probability)')
ax1.view_init(elev=15., azim=2)

# ----- Second subplot: 2D projection of the decision boundary -----
ax2 = fig.add_subplot(1, 2, 2)  # 1 row, 2 columns, second plot

# Plot the decision boundary line
ax2.plot(x, y_decision_boundary, color='black',linewidth=3, label='Decision Boundary')
ax2.fill_between(x, y_decision_boundary, y[0], color='blue', alpha=0.3, label = 'class 1')  # Optional: fill area under the line
ax2.fill_between(x, y_decision_boundary, y[-1], color='red', alpha=0.3, label = 'class 2')  # Optional: fill area under the line

ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Decision Boundary in XY plane', fontsize=14)
ax2.legend()

plt.tight_layout() 
plt.show()
```

## Task 4 reflections

**The decision boundary is defined by the plane where $z\in[0,1]$ and x,y such that $ b + w_1*x+w_2*y = 0 $ . Why is this?**

This gives a line, for which positive values (one side of the line), means they are predicted to be class1, and negative values (other side of the line), means they are predicted to be class2.

**How does changing the individual model parameters  $w_1$  and  $w_2$  affect the orientation and slope of the decision boundary plane in the 3D plot?**



Changing the coefficient for x ($w_1$) the decision boundary plane seems to rotate around the z-axis such that it becomes steeper in the x-direction when it increases less steep when it decreases. The same happens for $w_2$ in the y-direction.

**What happens to the decision boundary if  $w_1$  or  $w_2$  is set to zero?**

We more or less make it 2D, and only look at the variable which is non-zero and the bias, to calculate everything.

```python
def predict2D(X,params):
    """
    :param x: tuple of 1D arrays of the input-feature values.
    :param params: The list of the model parameters, [bias, weight1, weight2]. 
    
    :return: Boolean array same size as x, where a True values signifies class2, and False signifies class1
    """
    # Calculate the sigmoid output for each (x, y) pair using sigmoid2D
    sigmoid_output = sigmoid2D(X, params)
    
    # Apply threshold at 0.5 to classify as True (class2) or False (class1)
    return sigmoid_output >= 0.5


X_data = (np.concatenate([p1, p2]), np.concatenate([q1, q2]))  # Tuple of x and y coordinates
y_true = np.concatenate([z1, z2])  # Combined labels for both classes

#New params, with bias, x, y
params = [0.07,0.037,0.085]  # Example parameters [bias, weight_x, weight_y]

# Use predict2D to get predictions for the combined data
predictions = predict2D(X_data, params)

# Calculate accuracy by rounding y_true to integers and comparing with predictions
accuracy_score = accuracy(predictions, y_true.round())

# Output the accuracy
print(f'Accuracy of the classifier with parameters {params}: {accuracy_score:.3f}')
```
